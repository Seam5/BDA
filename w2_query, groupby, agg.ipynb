# 파생변수
- 기존의 변수 또는 컬럼 값을 가지고 새로운 변수를 만들 수 있음
- 파생변수는 새로운 변수지만 기존의 변수들의 관계를 가지고 만든다.
- assign.(만들 변수 = 식)


```python
exam.assign(total_3 = (exam['math']+exam['english']+exam['science'])/3,
            sum_3 = exam['math']+exam['english']+exam['science'])
```

-> 셀에서 작동이 끝나면 원본 데이터에는 반영되지 않음

```python
exam_1 = exam.assign(total_3 = (exam['math']+exam['english']+exam['science'])/3,
            sum_3 = exam['math']+exam['english']+exam['science'])
exam_1 # 새로운 변수에 저장하는 방식으로 원본 데이터에 추가할 수 있음
```

# groupby, agg
- 엑셀에서 말하는 피벗 개념
- 원하는 컬럼만 요약통계 볼 수 있음 (집단별 요약통계 출력)
- groupby : 집단을 만드는 함수
- agg : 다양한 통계치 사용할 수 있는 함수
ex. agg(매개변수 = ('컬럼명','원하는 요약통계')


```python
exam.groupby('nclass').agg(mean_english = ('english','mean')) #그룹별 english 평균
```

```python
exam.groupby('nclass').agg(sum_english = ('english','sum')) #그룹별 english 합계
```

```python
exam.groupby('nclass').agg(medi_english = ('english','median')) #그룹별 english 중앙값
```

```python
exam.groupby('nclass').agg(max_english = ('english','max')) #그룹별 english 최댓값
```

```python
# 여러 그룹별 요약통계치 한 번에 보여주는 코드
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                         sum_math = ('math','sum'),
                         max_math = ('math','max'),
                         n = ('nclass','count'), #개수 카운팅  
                         std_math = ('math','std')) # std : 표준편차 
```

# merge
- 데이터 병합
- sql 데이터 추출이 필요할 때 데이터가 다른 테이블에 있는 경우가 많고 두개의 df을 추출해서 하나로 합쳐야 하는 경우가 존재
ex. 공공데이터와 다른 데이터를 같이 사용하고 싶어 두 데이터를 붙여야 하는 경우
- 데이터를 합치기 위해서는 공통된 기준이 있어야 함

```python
a = pd.DataFrame({'name':['김철수','김영희','홍길동'],
                 '영어 점수':[10,20,30]})
b = pd.DataFrame({'name':['김철수','김영희','홍길길'],
                 '수학 점수':[10,20,40]})
```

```python
display(a)
display(b)
```

- merge를 이용하려면 공통된 기준이 필요.

```python
total = pd.merge(a,b, how='inner')
```

```python
total
```

- concat : 덩어리 + 덩어리로 붙는 형태
- 공통된 부분 없어도 데이터 프레임을 붙일 수 있음

```python
pd.concat([a,b]) #axis 축 default는 0 값
```

```python
pd.concat([a,b],axis=1) 
```

- axis = 0 : 열(세로)방향 계산
- axis = 1 : 행(가로)방향 계산

- merge, concat은 어떤 상황에서 사용하느냐에 따라 활용
- 두 함수의 원리를 정확하게 이해하고 있어야 대량의 데이터를 합칠 때 문제가 발생하지 않음.
