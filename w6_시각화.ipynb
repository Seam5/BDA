# 막대 그래프
- plt.bar(x 값 지정, y 값 지정)
- 눈금값 설정 : xticks(x값, 눈금값)
- 스타일 설정 : color, width(막대 두께), linewidth(선 두께)

```python
plt.bar(range(5),[10,20,30,40,50]) # bar : 막대그래프
```

```python
plt.bar(range(5),[10,20,30,40,50])
plt.xticks(range(5), ['a','b','c','d','e'])
```

# 히스토그램
- 도수분포표를 시각화 한 것
- 도수분포표 : hist
- 히스토그램 : bins를 통해 한 칸의 크기 조정 -> plt.hist(x, bins = n)
=> 칸의 개수가 많을 수록 정교화
- 옵션 : color, alpha(투명도) 지정 가능
- edgecolor : 각 구간을 구분하는 선 지정 가능
- cumulative = True => 누적 히스토그램
- histtype으로 스타일 설정 :  'barstacked', 'step', 'stepfilled' 

```python
x = np.random.normal(5,10,100)
y = np.random.normal(7,5,100)
```

```python
plt.hist(x)
```

```python
plt.hist(x, bins = 20) # 데이터 분포에 대한 왜곡 가능성이 있으므로 bins는 신중하게 선택 (15~20)
```

```python
plt.hist(x, bins = [-20,-15, -10, 0, 5, 10, 15, 20], edgecolor = 'black') #bins 인자로 구간 수동 설정도 가능
```

```python
plt.hist(x, edgecolor = 'black', range = (-15,15), histtype = 'step') # range 설정, histtype 설정
```

- 왼쪽에 이상치 추정 -> range로 범위 좁혀주기

```python
plt.hist(x, color = 'pink', bins = 30, range = (-15,15), cumulative = True)
```

```python
plt.hist(x, bins = 20, range = [-15,15], color = 'green', label = 'x', density = True) # density : y축에서 각 구간의 개수가 아닌 빈도가 표시됨 
plt.hist(y, bins = 30, color = 'skyblue', label = 'y', density = True) 
plt.legend() # 범례 추가
plt.show()
```

# 히트맵
- 2차원 배열로 각 요소의 색상을 시각화 하는 것
- 상관계수 표현할 때와 비슷
- annot : 박스 대응 데이터 표시
- sns.set(font_scale = 0.8) => 폰트 크기 설정
- fmt : 값의 형태 -> 'd'는 정수 형태
- cmap : 색상

```python
data = pd.DataFrame({'DA':[50,70,80], 'PG':[10,20,30]},
                    index = ['A','B','C'])
data
```

```python
sns.heatmap(data)
```

```python
sns.set(font_scale = 0.8)
sns.heatmap(data, annot = True, fmt = 'd', linewidth = 10, linecolor = 'white') # 선 두께, 색상 설정
```

```python
sns.heatmap(data, annot = True, fmt = 'd', linewidth = 10, linecolor = 'white', cmap = 'Greens') #cmap : 색상 설정(앞글자 대문자, 끝에 s붙임)
```

# 그래프 여러 개 그리기
- subplots을 통해 그래프를 여러 개 그릴 수 있음
1. fig : 도화지의 개념 그림의 밑바탕을 그리는 것
2. axis ax : 실제 그림 그리는 것
- fig, (데이터, 데이터) = plt.subplots(nrows = n, ncols = n, figsize = ())
- fig, 데이터 = plt.subplots(nrows = 1, ncols = 2, figsize = (10,8))
- 데이터[0].plot(mpg['cty'])
- 데이터[1].plot(mpg['model'])
=> 이런 식으로 데이터 하나만 지정해서 row, col 설정하는 방식으로 매핑 할 수도 있음.

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(5,5)) #ncols는 데이터 수에 맞게
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,6))
ax[0].plot(mpg['cty'])
ax[1].hist(mpg['model'].value_counts(), color = 'green')
```

- 반복문을 위해서는 도화지가 필요하고, 도화지를 몇 개로 할지 정해야 함
  - 도화지를 2x2 로 하고싶으면 4개의 그림을 그릴 준비 해야 함

# gridspec 라이브러리 활용

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) #그래프 4가지를 그리겠다
plt.figure(figsize = (16,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.4) #여러 개 그래프 그리기 때문에 충돌하지 않게 간격 조절

mpg_features = ['manufacturer','model','category','drv']
# 4가지 모두 문자열 데이터 -> count에 대한 시각화

for idx, feature in enumerate(mpg_features):

  ax = plt.subplot(grid[idx])

  sns.countplot(x = feature,
                data = mpg,
                hue = 'drv', #hue : ~에 따라 색을 다르게 하라
                palette = 'pastel', ax = ax) #색상 조절
  ax.set_title(f'{feature} Distribution') #data.set_title() : 제목 지정
```

### enumerate() : 인덱스와 원소로 이루어진 튜플을 생성
