# na값을 대체하는 방법
- fillna(method = "")
- method
- ffill(누락값이 나타나기 전의 값으로 누락값 대체)
- bfill(누락값이 나타난 이후 첫 번째 값으로 누락값 대체)

```python
ebola.iloc[0:10, 0:5]
```

```python
ebola.fillna(0).iloc[0:10, 0:5]['Cases_Liberia'].mean() # 0으로 결측치 대체 후 평균 = 4829.5
```

```python
ebola.fillna(method = "ffill").iloc[0:10, 0:5]['Cases_Liberia'] # 0,1의 결측치 안채워짐
```

```python
ebola.fillna(method = "bfill").iloc[0:10, 0:5]['Cases_Liberia'].mean()  # bfill 결측치 처리 후 평균 = 8046.6
```

# 보간법
- interpolate() : 시계열 데이터, 공간변수가 포함된 데이터 보간에 효과적

```python
ebola.interpolate().iloc[0:10, 0:5]['Cases_Liberia']
```

```python
ebola.interpolate(method = 'quadratic').iloc[0:10, 0:5]['Cases_Liberia'] #quadratic : 양쪽에 있는 값들의 중간값으로 누락값 대체
```

```python
ebola['Cases_Liberia'].plot(figsize = (5,5))
```

- 선이 매끄럽지 않고 끊겨있는 형태

```python
ebola.fillna(8049.2)['Cases_Liberia'].plot(figsize = (5,5)) # 평균 값으로 누락값 대체
```

- 부드럽지 않은 형태

```python
ebola.interpolate()['Cases_Liberia'].plot(figsize = (5,5))
```

## interpolate method
- ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']
- 다양한 보간법 사용 가능

```python
ebola.interpolate(method = 'linear')['Cases_Liberia'].plot(figsize = (5,5)) #linear : 선형보간법
```

- linear (선형보간법) : 직선거리에 따라 결측치 추정하는 방식

```python
ebola.interpolate(method = 'quadratic')['Cases_Liberia'].plot(figsize = (5,5)) #quadratic : 스플라인 보간법
```

- quadratic(스플라인 보간법) : 전체 구간을 소구간으로 나눈 뒤, 다항식에 의해 결측치를 추정하는 방식
- 다항식을 사용하기 때문에 선형보간법 보다 부드럽게 연결됨 (꺾이는 정도가 덜 함)

# melt 함수
- 항상 깔끔한 데이터만 있는 것은 아니며, 정제되지 않은 데이터 多
- 데이터 재구조화 방법 중 하나 : melt() 
- 판다스로 정의 : pd.melt(df, id_vars = '')
- 특정 colnum 기준으로 특정 값을 정렬


- id_vars : identy variables로 쓸 컬럼(위치를 그대로 유지할) 지정
- value_vars : 행으로 위치를 변경할 열의 이름 지정
- var_name : value_vars를 대표하는 이름 지정
- value_name : value 값 이름 지정


```python
pew = pd.read_csv('/content/pew.csv')
```

```python
pew
```

```python
pd.melt(pew, id_vars = 'religion')
```

```python
pd.melt(pew, id_vars = 'religion', var_name = 'income', value_name = 'count')
```

```python
bill = pd.read_csv('/content/billboard.csv')
```

# 문자열 분리
- 문자열을 분리하여 각각 다른 열에 저장하고 싶을 때
- .str.split() 사용 -> pandas의 series 형태로 저장,
- .str.get()을 통해 불러와서 저장.

```python
vr_sp = ebola_sp.variable.str.split('_') # '_'를 기준으로 변수 분리
vr_sp
```

```python
case = vr_sp.str.get(0)
case
```

```python
case_cr = vr_sp.str.get(1)
case_cr
```

```python
ebola_sp['case'] = case
ebola_sp['case_cr'] = case_cr
```

```python
ebola_sp
```
