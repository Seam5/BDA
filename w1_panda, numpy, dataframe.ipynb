#추가 공부 정리

### 데이터 입출력

- read_~ : 다양한 읽기 라이브러리가 제공됨  / csv, excel, html, json… 등
- df.read(2) : 2줄 읽기
- to_~ : 다양한 포맷으로 저장 가능 / ex. df = a.to_frame()

### 데이터 선택

- data[start : end-1] : start 부터 end-1까지 추출
- data[key1, key2, key3….] : 특정 key data만 추출한다

### 판다스
- 엑셀처럼 data 형태가 관리 됨
- 사용 필수 조건 : import pandas as pd
- 데이터분석을 위한 라이브러

### plot
- plot() : 기본 설정인 꺾은선(line) 그래프
- 특정 필드 값만 가져와서 그래프 만들고 싶다면 파라미터 값으로 y축에 원하는 필드명 지정
- 그래프 상세 설정하기
    - y축 로그(log) 스케일 : logy=True
    - y축 최대, 최솟값 설정 : ylim=(100,500)
    - 그래프 그리드 생성 : grid = True
    - x, y축 레이블(이름) 표시 : xlable = “x축”, ylable = “y축”
    - 그래프 전체 크기 설정 : figsize(10,6)
    - 마커 표시 및 마커 크기 설정 : marker=’o’, ms=3
    - 라인 굵기 : lw =1


### 넘파이
- np.where(condition, T, F) 값 변경 : 조건 만족하면 True, 그렇지 않으면 False 적용
- np.where() - 인덱스 반환, 값 반환
- 조건에 맞는 값 찾기 - np.max, np.min, np.sort
- 조건에 맞는 인덱스 찾기 - np.argmax, np.argmin, np.argsort

```python
import numpy as np
```

```python
a = np.arange(10,20)
```

```python
np.where(a>15, True, False)
```

```python
np.where(a>15)  #a가 15보다 큰 경우의 인덱스 반환
```

```python
np.max(a) #최댓값 반환
```

```python
np.sort(a) #작은 값부터 순서대로 반환
```
